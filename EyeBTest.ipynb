```python
import numpy as np
import cv2
import tensorflow.compat.v1 as tf
import os
import matplotlib.pyplot as plt
tf.disable_v2_behavior() # Disable changes made in tf version 2.x
```

```python
CATEGORIES = ["closed", "open"]
IMG_DIM = 50
MODELS_FOLDER_PATH = os.path.join(os.getcwd(), "models")
TEST_FOLDER_PATH = os.path.join(os.getcwd(), "test_data")
```

```python
def select_model():
    index = 0
    for model_name in os.listdir(MODELS_FOLDER_PATH):
        print(index+1, ". " + model_name)
        index += 1
    index = int(input("Enter model number: ")) - 1
    model_name = os.listdir(MODELS_FOLDER_PATH)[index]
    model = tf.keras.models.load_model(os.path.join(MODELS_FOLDER_PATH, model_name))
    return model
```

```python
def select_test_image():
    index = 0
    for test_image_name in os.listdir(TEST_FOLDER_PATH):
        print(index+1, ". " + test_image_name)
        index += 1
    index = int(input("Enter image number: ")) - 1
    test_image_name = os.listdir(TEST_FOLDER_PATH)[index]
    test_image_path = os.path.join(TEST_FOLDER_PATH, test_image_name)
    return test_image_path
```

```python
def test_model(model, test_image_path):
    img = cv2.imread(test_image_path, cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(img, (IMG_DIM, IMG_DIM))
    image = image.reshape(-1, IMG_DIM, IMG_DIM, 1)
    prediction = model.predict([image])
    print("Eye is: ", CATEGORIES[int(prediction[0][0])])
    plt.imshow(img, cmap='gray')
    plt.show()
    
```

```python
model = select_model()
```

```python
image_path = select_test_image()
```

```python
test_model(model, image_path)
```
